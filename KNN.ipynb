```python
import numpy as np
from collections import Counter
```

```python
class KNNClassifier:
    def __init__(self, k=3):
        self.k = k
        self.X_train = None
        self.y_train = None

    def fit(self, X, y):
        self.X_train = np.array(X)
        self.y_train = np.array(y)

    def predict(self, X):
        X = np.array(X)
        return np.array([self._predict_single(x) for x in X])

    def score(self, X, y):
        y_pred = self.predict(X)
        return np.mean(y_pred == np.array(y))

    def _predict_single(self, x):
        distances = [self._euclidean_distance(x, x_train) for x_train in self.X_train]
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = self.y_train[k_indices]
        return Counter(k_nearest_labels).most_common(1)[0][0]

    def _euclidean_distance(self, x1, x2):
        return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
X = np.array([
    [2.7, 2.5],
    [1.3, 1.8],
    [3.0, 3.2],
    [1.2, 0.9],
    [3.1, 2.9],
    [1.0, 1.1],
    [2.9, 3.0],
    [1.4, 1.3]
])

y = np.array([1, 0, 1, 0, 1, 0, 1, 0])
```

```python
knn = KNNClassifier(k=3)
knn.fit(X, y)

predictions = knn.predict(X)
accuracy = knn.score(X, y)

print("Predictions:", predictions)
print("Accuracy:", accuracy)
```

```python
class KNNClassifierWithProba:
    def __init__(self, k=3):
        self.k = k
        self.X_train = None
        self.y_train = None
        self.classes_ = None

    def fit(self, X, y):
        self.X_train = np.array(X)
        self.y_train = np.array(y)
        self.classes_ = np.unique(y)

    def predict(self, X):
        X = np.array(X)
        return np.array([self._predict_single(x) for x in X])

    def predict_proba(self, X):
        X = np.array(X)
        return np.array([self._predict_proba_single(x) for x in X])

    def score(self, X, y):
        y_pred = self.predict(X)
        return np.mean(y_pred == np.array(y))

    def _predict_single(self, x):
        distances = np.sqrt(np.sum((self.X_train - x) ** 2, axis=1))
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = self.y_train[k_indices]
        return Counter(k_nearest_labels).most_common(1)[0][0]

    def _predict_proba_single(self, x):
        distances = np.sqrt(np.sum((self.X_train - x) ** 2, axis=1))
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = self.y_train[k_indices]
        counts = np.array([np.sum(k_nearest_labels == c) for c in self.classes_])
        return counts / self.k
```
